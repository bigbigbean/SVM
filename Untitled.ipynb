导入模块

```python
import numpy as np
from scipy import io as spio
from matplotlib import pyplot as plt
from sklearn import svm
```

定义SVM的流程：
1. 用spio来载入mat格式的数据存储在data里
2. 将X数据和y数据分开存储在两个向量格式的数据里
3. y=np.ravel(y)不知道什么意思
4. 将X，y的数据可视化（流程见下）
5. 用支持向量机模型（线性核函数或高斯核函数）去fit数据，训练模型
6. 画出训练完成的决策边界（流程见下）

```python
def SVM():
    '''data1——线性分类'''
    data1 = spio.loadmat('data1.mat')
    X = data1['X']
    y = data1['y']
    y = np.ravel(y)
    plot_data(X,y)
    
    model = svm.SVC(C=1.0,kernel='linear').fit(X,y) # 指定核函数为线性核函数
    plot_decisionBoundary(X, y, model)  # 画决策边界
    '''data2——非线性分类'''
    data2 = spio.loadmat('data2.mat')
    X = data2['X']
    y = data2['y']
    y = np.ravel(y)
    plt = plot_data(X,y)
    plt.show()
    
    model = svm.SVC(gamma=100).fit(X,y)     # gamma为核函数的系数，值越大拟合的越好
    plot_decisionBoundary(X, y, model,class_='notLinear')   # 画决策边界
```

数据可视化流程：
1. 定义fig的size为10，8
2. 将y=1，y=0的点分开存储在pos和neg变量中
3. 先将pos的点可视化，再将neg的点画在同一幅图上
4. 在图上标好x，y轴的标签和legend

```python
# 作图
def plot_data(X,y):
    plt.figure(figsize=(10,8))
    pos = np.where(y==1)    # 找到y=1的位置
    neg = np.where(y==0)    # 找到y=0的位置
    p1, = plt.plot(np.ravel(X[pos,0]),np.ravel(X[pos,1]),'ro',markersize=8)
    p2, = plt.plot(np.ravel(X[neg,0]),np.ravel(X[neg,1]),'g^',markersize=8)
    plt.xlabel("X1")
    plt.ylabel("X2")
    plt.legend([p1,p2],["y==1","y==0"])
    return plt
```

画决策边界的步骤：
1. 定义plotboundary的函数，入口需要提供的变量有X，y，model和是否线性可分
2. 得到数据后先把X，y的数据可视化，在图上分开pos和neg，即上面的步骤
3. 如果是线性可分的数据，w=训练得到的model的斜率，b=训练得到的model的截距（与y轴的交点）
   然后画线，xp和yp（不太懂）
4. 如果是线性不可分的数据，transpose？meshgrid?不明白？？

```python
# 画决策边界
def plot_decisionBoundary(X,y,model,class_='linear'):
    plt = plot_data(X, y)
    
    # 线性边界        
    if class_=='linear':
        w = model.coef_
        b = model.intercept_
        xp = np.linspace(np.min(X[:,0]),np.max(X[:,1]),100)
        yp = -(w[0,0]*xp+b)/w[0,1]
        plt.plot(xp,yp,'b-',linewidth=2.0)
        plt.show()
    else:  # 非线性边界
        x_1 = np.transpose(np.linspace(np.min(X[:,0]),np.max(X[:,0]),100).reshape(1,-1))
        x_2 = np.transpose(np.linspace(np.min(X[:,1]),np.max(X[:,1]),100).reshape(1,-1))
        X1,X2 = np.meshgrid(x_1,x_2)
        vals = np.zeros(X1.shape)
        for i in range(X1.shape[1]):
            this_X = np.hstack((X1[:,i].reshape(-1,1),X2[:,i].reshape(-1,1)))
            vals[:,i] = model.predict(this_X)
        
        plt.contour(X1,X2,vals,[0,1],color='blue')
        plt.show()
```

运行函数SVM，观察线性可分的和不可分的图像以及model边界

```python
if __name__ == "__main__":
    SVM()
```
